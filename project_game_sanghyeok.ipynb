```python
import random

class Character:
    print('운빨 디지털 세계에 오신 것을 환영합니다.')
    print('당신은 랜덤으로 공격값이 나타나는 주사위 폭탄을 들고 몬스터를 사냥할 것입니다.')
    print('주사위 폭탄을 사용하여 시간의 마법사를 저지하는 모험을 떠나봅시다.')
    name = input('캐릭터 이름을 입력해 주세요: ')

    print("      _______")
    print("     /      /|")
    print("    /  ●  / |")
    print("   /______/  |")
    print("  | ● ●|● |")
    print("  |  ●  | ●|")
    print("  |●__●|_ /")
    print(' ')
    print('가상의 주사위를 굴려 체력, 공격력, 방어력을 결정합니다.')

    life = []
    attack = []
    defense = []
    power = 0

    while power == 0:
        life.append(random.randint(1, 10))
        attack.append(random.randint(1, 10))
        defense.append(random.randint(1, 10))

        print(f'체력: {life}, 공격력: {attack}, 방어력: {defense}')
        power = int(input('재시도? 재시도는 0, 확정은 1을 입력하라. '))
        if power == 1:
            print('완료')
        else:
            life.pop()
            attack.pop()
            defense.pop()

    def __init__(self):
        self.life = Character.life[-1]  # 최신 체력 값 적용
        self.attack = Character.attack[-1]  # 최신 공격력 값 적용
        self.defense = Character.defense[-1]  # 최신 방어력 값 적용
        self.name = Character.name  # 이름

    def is_alive(self):
        return self.life > 0  # 체력이 0보다 크면 살아 있음


    def attack_target(self):
        attack_value = random.randint(1, self.attack)  # Generate random attack value
        print(f'{self.name}가 던진 주사위 폭탄이 {attack_value}의 피해를 입힙니다!')
        return attack_value  # Return the attack value

    def take_damage(self, damage):
        if damage - self.defense < 0:
            print(f'당신의 엄청난 방어력에 몬스터가 오히려 피해를 입었습니다!!!')
            counter_attack = -damage + self.defense
            print(f'{self.name}이(가) {counter_attack} 피해를 주었습니다.')
        else:
            damage_taken = damage - self.defense  # Reduce damage considering defense
            self.life -= damage_taken
            print(f'{self.name}이(가) {damage_taken} 피해를 입었습니다. 남은 체력: {self.life}')
```

```python
class Monster(Character):  # Character 클래스 그대로 유지
    def __init__(self, name, level):
        super().__init__()
        self.name = name  # name을 직접 설정하도록 수정
        self.level = level
        self.life = random.randint(10, 30) * level
        self.attack = random.randint(5, 20) * level
        self.defense = random.randint(1, 5) * level
        print(f'몬스터 등장! {self.name} - 레벨: {self.level}, 체력: {self.life}, 공격력: {self.attack}, 방어력: {self.defense}')
```

```python
class Player(Character):  # Character 클래스를 건드리지 않음
    def __init__(self):
        super().__init__()
        self.level = 1
        self.life += 100
        self.attack += 25
        self.defense += 5
        self.experience = 0

    def gain_experience(self, exp):
        self.experience += exp
        print(f'{self.name}이(가) {exp} 경험치를 획득했다! 현재 경험치: {self.experience}')
        while self.experience >= 50:
            self.experience -= 50
            self.level_up()

    def level_up(self):
        self.level += 1
        self.attack += 50
        self.defense += 5
        print(f'{self.name}이(가) 레벨 업! 현재 레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense}')

    def item(self):
        self.attack += random.randint(-10, 10)
        self.life += random.randint(-20, 20)
        self.defense += random.randint(-5, 5)
        print(f'아이템 사용! 체력: {self.life}, 공격력: {self.attack}, 방어력: {self.defense}')
```
